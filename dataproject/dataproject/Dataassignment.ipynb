```python
def prices(name, start=start, end=end):
    '''returns a dataframe with stock-information for a given company'''
    return pandas_datareader.iex.daily.IEXDailyReader(name, start, end).read()
#creats a function called "Prices" where the parameter "name" is the company acronym.
```

```python
# Now we create a Keep function, where we can pinpoint which columns we want to keep
```

```python
def keep(Prices, name, open = "Remove", high = "Remove", low = "Remove", close = "Remove" , volume = "Remove" ):
    keep_list = [open, high, low, close, volume]
    prices = Prices(name).loc[:,[open, high, low, close, volume]]
    Cleandataframe = prices.drop(columns= "Remove")
    return Cleandataframe
```

```python
Keep(Prices, "IBM", "open", close="close").head(10) #we are primarily interested in the open and close price
```

```python
Stock = Keep(Prices, "IBM", "open", close="close")
```

```python
Stock
```

```python
#Now we find some google trends data
```

```python
def Searches(*kw_list):
    pytrends.build_payload(kw_list, cat=0, timeframe='2019-2-1 2019-3-18', geo='', gprop='')
    searches = pytrends.interest_over_time()
    trend = searches.drop(columns = "isPartial")
    #sørg for at den kommer til at hedde kw_list + searches, eksempelvis hvis kw_list = "IBM" så returner den IBMsearches 
    return trend
```

```python
Popularity = Searches("IBM")
```
